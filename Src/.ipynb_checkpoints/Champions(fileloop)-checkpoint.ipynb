```python
import pandas as pd
import numpy as np
import re
import mysql.connector as conn
from sqlalchemy import create_engine
import os #modulo para acceso a los ficheros
```

```python
def quita_porcentajes(x):
    return x[:-1]
```

```python
st12_20=pd.read_csv(r'C:\Users\david\OneDrive\Escritorio\Proyectos\Mad Scientist\datos\campeones\League of Legends Champion Stats 12.20.csv', sep=';')
```

```python
st12_20 = st12_20.rename(columns={'Win %': 'win', 'Role %': 'pick_role', 'Pick %':'pick', 'Ban %':'ban'})
```

```python
#eliminacion del simbolo % en los datos
st12_20['win']=st12_20.win.apply(quita_porcentajes)
st12_20['pick_role']=st12_20.pick_role.apply(quita_porcentajes)
st12_20['pick']=st12_20.pick.apply(quita_porcentajes)
st12_20['ban']=st12_20.ban.apply(quita_porcentajes)
#trasformacioin del dato a numerico
st12_20['win'] = st12_20['win'].astype(float)
st12_20['pick_role'] = st12_20['pick_role'].astype(float)
st12_20['pick'] = st12_20['pick'].astype(float)
st12_20['ban'] = st12_20['ban'].astype(float)
```
