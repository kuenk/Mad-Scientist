Scrapeo de personajes para obtener datos porcentuales de victorias y derrotas globales para el rango de Challenger

```python
URL='https://www.op.gg/champions'
```

```python
from webdriver_manager.chrome import ChromeDriverManager
PATH=ChromeDriverManager().install()
driver=webdriver.Chrome(PATH, options=opciones)
driver.get(URL)
time.sleep(5)

try:
    texto=driver.find_element(By.XPATH, '//*[@id="qc-cmp2-ui"]/div[2]/div/button[2]') #aceptamos cookies
    texto.click()
    time.sleep(5)
except:
    pass

texto=driver.find_element(By.XPATH, '//*[@id="content-container"]/div[1]/nav[1]/div[4]/div/button')
texto.click()
time.sleep(2)
texto=driver.find_element(By.XPATH, '//*[@id="content-container"]/div[1]/nav[1]/div[4]/div/div/button[2]')
texto.click()
time.sleep(2)

#recuperacion del link de cada uno de los invocadores de la lista
champlinks=[]
for e in driver.find_element(By.CLASS_NAME,'css-1x3kezq').find_elements(By.TAG_NAME,'li'):
    champlinks.append(e.find_element(By.TAG_NAME, 'a').get_attribute('href'))
    
    
driver.quit()
```

```python
#funcion que devuelve las estadisticas de las partidas por jugador
def campeones(link):
    driver=webdriver.Chrome(PATH, options=opciones)
    driver.get(link)
    time.sleep(5)

    try:
        texto=driver.find_element(By.XPATH, '//*[@id="qc-cmp2-ui"]/div[2]/div/button[2]') #aceptamos cookies
        texto.click()
        time.sleep(2)
    except:
        pass

    #recuperacion del link de cada uno de los invocadores de la lista
    res=[]
    try:
        for e in driver.find_element(By.CLASS_NAME, 'css-c3v1ys.ew1oorz8').find_elements(By.CLASS_NAME, 'css-kjoz7i.ew1oorz7'):
            res.append(e.find_element(By.CLASS_NAME, 'css-oxevym.ew1oorz5').text)
    except:
        nulos=('0%','0%','0%')
        res.append(nulos)
    time.sleep(1)
    driver.quit() 
    
    return res
```

```python
porcentajes=Parallel(n_jobs=2, verbose=True)(delayed(campeones)(url) for url in champlinks)
```

```python
porcen=[]
for e in porcentajes:
    if isinstance(e[0], list):
        for lst in e:
            porcen.append(lst)
    else:
        porcen.append(e)
     
```

```python
columna=['win','pick','ban']
```

```python
campeones=pd.DataFrame(porcen, columns=columna)
```

```python
campeones.to_csv('../datos/campeonesjson.csv', sep=';')
```
